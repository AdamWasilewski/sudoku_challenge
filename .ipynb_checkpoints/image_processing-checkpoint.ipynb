```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
img = cv2.imread('sample_images\\sudoku1.jpg')
```

```python
matplotlib.rcParams['figure.figsize'] = [10,10]
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.GaussianBlur(img,(25,25),0)
```

```python
img = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,21,2)
img = cv2.bitwise_not(img)
# plt.imshow(img, cmap='gray')
# plt.show()
```

```python
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, (25,25))
dilation = cv2.dilate(opening,(5,5),iterations = 3)
blur = cv2.GaussianBlur(dilation,(25,25),0)
dilation = cv2.dilate(blur,(5,5),iterations = 3)
blur = cv2.GaussianBlur(dilation,(25,25),0)
dilation = cv2.dilate(blur,(5,5),iterations = 3)
blur = cv2.GaussianBlur(dilation,(25,25),0)
plt.imshow(blur, cmap='gray')
plt.show()
```
